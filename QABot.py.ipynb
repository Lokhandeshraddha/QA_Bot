Import required libraries

```python
import gradio as gr
from transformers import pipeline
```

Load QA Pipeline

```python
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

Titanic Dataset Example

```python
context = """
The Titanic was a British passenger ship that sank in 1912 after hitting an iceberg.
It carried 2,224 passengers and crew, but only around 710 survived.
Survival depended on factors like passenger class, gender, and age.
Women and children had higher survival chances compared to men.
"""
```

# Gradio App

```python
def answer_question(user_q):
    if user_q.lower() == "quit":
        return "Goodbye 👋"
    answer = qa_pipeline(question=user_q, context=context)
    return answer['answer']

iface = gr.Interface(
    fn=answer_question,
    inputs="text",
    outputs="text",
    title="LLM Q&A Bot",
    description="Ask me questions based on given context."
)

iface.launch(share=True)
```
